```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.utils.data
import torchvision.models as models
from collections import OrderedDict
from torch.autograd import Variable

import time
import pickle
import pandas
import numpy as np 
import matplotlib.pyplot as plt
import scipy.stats as st
import gc
```

```python

# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - 
#A) main functions definition
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - 


def ResNet_18(output_classes=1):
  """
    Load a pretrained ResNet-18 model from PyTorch with a specific last dense layer with a dedicated number of neurons.
    Previous layers are already frozen to update and train only the last classification layer.
    
    :param output_classes: The number of output classes in our classification problem, defining the number of neurons in the last layer.
    
    :type output_classes: integer
  """
  # Load a pretrained ResNet-18 model thanks to PyTorch
  resnet_model = models.resnet18(pretrained = True)
  # Change the last dense layer to fit our problematic needs
  resnet_model.fc = nn.Sequential(OrderedDict([
    ('fc1', nn.Linear(512, 128)), ('relu', nn.ReLU()),
    ('fc2', nn.Linear(128, output_classes)), ('output', nn.Sigmoid())
  ]))
  
  return resnet_model



def LargeDatasetPred(model,var_X,BlockSizes):
  """
  Prediction of large datasets (incrementally predicts with 'model' blocks of observations out of 'var_X' having a size of 'BlockSizes').
  To be used in case the RAM is not large enough to treat all information in the NN. 
  """
  n_loc=var_X.shape[0]
  
  loc_miniBatch_Start=0
  
  while loc_miniBatch_Start<n_loc:
    #define the mini-batch domain
    loc_miniBatch_End=loc_miniBatch_Start+BlockSizes
    if loc_miniBatch_End >= n_loc:
      loc_miniBatch_End = n_loc
    
    #local prediction
    with torch.no_grad():
      loc_predY=model(var_X[loc_miniBatch_Start:loc_miniBatch_End,:,:,:])
    
    #merge local prediction with former ones
    if loc_miniBatch_Start==0:
      all_predY=torch.clone(loc_predY)
    else:
      all_predY=torch.cat([all_predY,loc_predY],dim=0)
    
    #increment loc_miniBatch_Start
    loc_miniBatch_Start+=BlockSizes
  
  return all_predY

def showRGBImage(LodID,X,Y,S):
  """
  show observations
  """
  LocImage=(X[LodID,:,:,:]*255).astype(int)
  LocTitle='Y='+str(int(Y[LodID,0]))+'    S='+str(int(S[LodID,0]))
  plt.figure()
  plt.imshow(LocImage)
  plt.title(LocTitle)
  plt.savefig('Observation_'+str(LodID)+'.png')
  plt.show()


  


def fit_model(model,X_data,y_data, EPOCHS = 5, BATCH_SIZE = 32):
    """
    -> The input observations (X_data) are supposed to have the shape, i.e. [batchSize,NbChannels,ImSizeX,ImSizeY]. 
       No 'view' is performed to adapt the data to the NN model
    -> The output observations (y_data) are a vector of binary values
    """

    optimizer = torch.optim.Adam(model.parameters())#,lr=0.001, betas=(0.9,0.999))
    error = nn.MSELoss()  # standard loss
    #error = nn.BCELoss()  #alternative standard loss

    model.train()
    
    n=X_data.shape[0]
    
    
    epoch=0
    while epoch<EPOCHS:
        obsIDs=np.arange(X_data.shape[0]) 
        np.random.shuffle(obsIDs)
        
        batch_start=0 
        batchNb=0
        
        
        while batch_start+BATCH_SIZE < n:
            print('Epoch '+str(epoch)+'  -- Batch start='+str(batch_start))
            
            #1) get mini-batch information
            Curr_obsIDs=obsIDs[batch_start:batch_start+BATCH_SIZE]
            
            Curr_obsIDs=obsIDs[batch_start:batch_start+BATCH_SIZE]
            
            var_X_batch = Variable(X_data[Curr_obsIDs,:,:,:]).float()
            var_y_batch = Variable(y_data[Curr_obsIDs,0].reshape(-1,1)).float()  #added in the 1d case
            
            optimizer.zero_grad()
            
            #2) prediction in the mini-batch
            output = model(var_X_batch)
            
                
            #4) compute the loss and perform the gradient descent step 
            loss = error(output, var_y_batch.view(-1,1))
            print(' -> loss='+str(loss))
            loss.backward()
            optimizer.step()
            
            #5) update the first observation of the batch
            batch_start+=BATCH_SIZE
            batchNb+=1
            
            #locOutput=1*(output.view(-1).data.numpy()>0.5)
            #locTruePred=1*(var_y_batch.view(-1).data.numpy()>0.5)
        
        epoch+=1
    
    torch.save(model.state_dict(), "./modelFinal.pytorchModel")         #save a model

    
```

```python
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - 
#B) get and format input data
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - 

Y_label='Eyeglasses'   #IN THE PAPER  -- others: 'Attractive', ...
S_label='Young'         #IN THE PAPER  -- others: 'Male', 'Pale_Skin', ...


#training set
infile = open('./DATA_celebA/train_64x64.pkl','rb')
X_train = pickle.load(infile)
infile.close()

Info_all_train=pandas.read_csv('./DATA_celebA/train.csv')

Y_train=(Info_all_train[Y_label].values).astype(np.float32).reshape(-1,1) 
S_train=(Info_all_train[S_label].values).astype(np.int32).reshape(-1,1)

#test set
infile = open('./DATA_celebA/test_64x64.pkl','rb')
X_test = pickle.load(infile)
infile.close()

Info_all_test=pandas.read_csv('./DATA_celebA/test.csv')
Y_test=(Info_all_test[Y_label].values).astype(np.float32).reshape(-1,1) 
S_test=(Info_all_test[S_label].values).astype(np.int32).reshape(-1,1)

##check the datasets
for LocID in range(0,10):
  print(LocID)
  showRGBImage(LocID,X_train,Y_train,S_train)
```

```python
#convert to torch format
torch_X_train = torch.from_numpy(X_train[:,:,:,:]).type(torch.FloatTensor).transpose(1,3).to(DEVICE)
torch_y_train = torch.from_numpy(Y_train[:,0].reshape(-1,1)).type(torch.FloatTensor).to(DEVICE)

torch_X_test = torch.from_numpy(X_test[:,:,:,:]).type(torch.FloatTensor).transpose(1,3).to(DEVICE)
torch_y_test = torch.from_numpy(Y_test[:,0]).type(torch.FloatTensor).to(DEVICE)
```

```python
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - 
#C) training phase
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - 


model = ResNet_18(1).to(DEVICE)


EPOCHS_in =5
BATCH_SIZE_in = 1000
Lists_Results=fit_model(model,torch_X_train,torch_y_train,  EPOCHS = EPOCHS_in, BATCH_SIZE = BATCH_SIZE_in)
```
